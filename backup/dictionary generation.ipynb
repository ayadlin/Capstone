```python
def split_hyphen_word(word):
    """ return a list of strings"""
    if '-' not in word:
        return [word]
    if len(word)-1 > word.index('-'):
        if not((word[word.index('-')+1]).isdigit()):
            return word.split('-')
        else:
            return [word]
    else:
        return [word[:-1]]
    
```

```python
def split_hyphen_words(sent):
    output = []
    for word in sent:
        output.extend(split_hyphen_word(word))
    return output
```
