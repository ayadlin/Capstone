```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/ale/Desktop/genes.csv')
```

```python
df = df.fillna('')
df.drop('Entrez Gene ID',axis=1, inplace=True)
df.drop('Entrez Gene ID(supplied by NCBI)', axis=1, inplace=True)
```

```python
dict_df = pd.DataFrame()
```

```python
df.dtypes
```

```python
dict_df['key'] = (df['Approved Name'].fillna('').str.cat(df['Previous Symbols']).str.cat(
    df['Previous Name']).str.cat(df['Synonyms']).str.cat(df['Name Synonyms']).str.cat(
    df['Accession Numbers']))
```

```python
dict_df.iloc[1][0]
```

```python
dict_df.head()
```

```python
dict_df['value'] = df['Approved Symbol']
```

```python
dict_df.head()
```

```python
gene_dict ={}
for row in dict_df.iterrows():
    name_list = row[1]['key'].replace('"', '').split(', ')
    main_name = row[1]['value']
    for name in name_list:
        gene_dict[name]=main_name
```
